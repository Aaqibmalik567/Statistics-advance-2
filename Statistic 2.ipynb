

1. **What is hypothesis testing in statistics?**
   Hypothesis testing is a statistical method used to make decisions or inferences about population parameters based on sample data. It involves formulating a null hypothesis and an alternative hypothesis, then using sample data to test which hypothesis is more likely to be true.

2. **What is the null hypothesis, and how does it differ from the alternative hypothesis?**
   The **null hypothesis (H₀)** is a statement that there is no effect or no difference. The **alternative hypothesis (H₁ or Ha)** is what you want to prove — it suggests that there is an effect or a difference. Hypothesis testing evaluates whether to reject H₀ in favor of H₁ based on evidence from data.

3. **What is the significance level in hypothesis testing, and why is it important?**
   The significance level (denoted as **α**) is the threshold for rejecting the null hypothesis. Common values are 0.05, 0.01, etc. It represents the probability of committing a **Type I error**, i.e., rejecting H₀ when it is actually true. It defines the sensitivity of the test.

4. **What does a P-value represent in hypothesis testing?**
   A **P-value** is the probability of obtaining results at least as extreme as those observed during the test, assuming the null hypothesis is true. It helps determine whether the observed data is consistent with H₀.

5. **How do you interpret the P-value in hypothesis testing?**

   * If **P-value ≤ α**: Reject the null hypothesis (evidence is statistically significant).
   * If **P-value > α**: Do not reject the null hypothesis (not enough evidence against H₀).

6. **What are Type 1 and Type 2 errors in hypothesis testing?**

   * **Type I Error**: Rejecting a true null hypothesis (false positive).
   * **Type II Error**: Failing to reject a false null hypothesis (false negative).

7. **What is the difference between a one-tailed and a two-tailed test in hypothesis testing?**

   * **One-tailed test**: Tests for deviation in only one direction (greater or lesser).
   * **Two-tailed test**: Tests for deviation in both directions (any difference).

8. **What is the Z-test, and when is it used in hypothesis testing?**
   A **Z-test** is used to determine whether there is a significant difference between sample and population means when the population standard deviation is known and the sample size is large (n ≥ 30).

9. **How do you calculate the Z-score, and what does it represent in hypothesis testing?**
   Z-score = $(X - μ) / σ$
   It measures how many standard deviations a data point (X) is from the population mean (μ). In testing, it helps assess how extreme the test statistic is under H₀.

10. **What is the T-distribution, and when should it be used instead of the normal distribution?**
    The **T-distribution** is used when the sample size is small (n < 30) and/or population standard deviation is unknown. It is wider and has heavier tails than the normal distribution.

11. **What is the difference between a Z-test and a T-test?**

    * **Z-test**: Population standard deviation known; large sample size
    * **T-test**: Population standard deviation unknown; small sample size

12. **What is the T-test, and how is it used in hypothesis testing?**
    A **T-test** assesses whether the means of two groups are statistically different. Types include one-sample, two-sample (independent), and paired T-tests.

13. **What is the relationship between Z-test and T-test in hypothesis testing?**
    Both compare sample statistics to population values. As sample size increases, the T-distribution approximates the Z-distribution, making both tests produce similar results for large samples.

14. **What is a confidence interval, and how is it used to interpret statistical results?**
    A **confidence interval** gives a range of values that is likely to contain the population parameter. It helps express uncertainty in estimates, commonly at 95% or 99% levels.

15. **What is the margin of error, and how does it affect the confidence interval?**
    The **margin of error** indicates the range above and below the estimate within which the true population value is expected. Larger margins yield wider intervals, reducing precision.

16. **How is Bayes' Theorem used in statistics, and what is its significance?**
    Bayes' Theorem allows updating the probability of a hypothesis based on new evidence. It’s significant in probabilistic modeling, diagnostics, and decision-making under uncertainty.

17. **What is the Chi-square distribution, and when is it used?**
    The **Chi-square distribution** is used to analyze categorical data. It is applied in tests for independence, goodness-of-fit, and variance.

18. **What is the Chi-square goodness of fit test, and how is it applied?**
    This test determines whether a sample matches a population with a specified distribution. It compares observed and expected frequencies.

19. **What is the F-distribution, and when is it used in hypothesis testing?**
    The **F-distribution** is used to compare two variances or in ANOVA testing. It’s right-skewed and dependent on two degrees of freedom (numerator and denominator).

20. **What is an ANOVA test, and what are its assumptions?**
    **ANOVA (Analysis of Variance)** tests whether there are significant differences among group means. Assumptions: independence, normality, and equal variances.

21. **What are the different types of ANOVA tests?**

    * **One-way ANOVA**: One independent variable
    * **Two-way ANOVA**: Two independent variables
    * **Repeated measures ANOVA**: Same subjects measured multiple times

22. **What is the F-test, and how does it relate to hypothesis testing?**
    The **F-test** compares variances to test if they are equal. It's used in ANOVA and regression to assess overall significance.

```python
# 3. Write a Python function to calculate the probability distribution function (PDF) of a Bernoulli distribution
def bernoulli_pmf(x, p):
    """
    x: 0 or 1 (possible values of Bernoulli)
    p: probability of success (1)
    returns probability mass function value
    """
    if x not in [0, 1]:
        return 0
    return p**x * (1-p)**(1-x)

print("P(X=1) for p=0.3:", bernoulli_pmf(1, 0.3))
print("P(X=0) for p=0.3:", bernoulli_pmf(0, 0.3))
```

```python
# Q14: Write a Python program to calculate the Z-score for a given data point and compare it to a standard normal distribution
import numpy as np
from scipy.stats import norm

def calculate_z_score(data_point, data):
    mean = np.mean(data)
    std = np.std(data)
    z = (data_point - mean) / std
    p_value = norm.cdf(z)  # cumulative probability for that z-score
    return z, p_value

data = np.random.normal(50, 10, 100)
data_point = 65
z, p = calculate_z_score(data_point, data)
print(f"Z-score: {z:.2f}, CDF value (probability): {p:.4f}")
```

```python
# Q15: Implement hypothesis testing using Z-statistics for a sample dataset
import numpy as np
from scipy.stats import norm

def z_test(sample, mu0, sigma):
    n = len(sample)
    sample_mean = np.mean(sample)
    z_stat = (sample_mean - mu0) / (sigma / np.sqrt(n))
    p_value = 2 * (1 - norm.cdf(abs(z_stat)))  # two-tailed test
    return z_stat, p_value

np.random.seed(0)
sample_data = np.random.normal(100, 15, 30)
mu0 = 110  # hypothesized population mean
sigma = 15  # population std dev known

z_stat, p_val = z_test(sample_data, mu0, sigma)
print(f"Z-statistic: {z_stat:.3f}, p-value: {p_val:.4f}")
if p_val < 0.05:
    print("Reject the null hypothesis")
else:
    print("Fail to reject the null hypothesis")
```

```python
# Q16: Create a confidence interval for a dataset using Python and interpret the result
import numpy as np
import scipy.stats as stats

def confidence_interval(data, confidence=0.95):
    n = len(data)
    mean = np.mean(data)
    sem = stats.sem(data)
    h = sem * stats.t.ppf((1 + confidence) / 2., n-1)
    return mean - h, mean + h

data = np.random.normal(50, 5, 100)
ci_low, ci_high = confidence_interval(data)
print(f"95% confidence interval: ({ci_low:.2f}, {ci_high:.2f})")
print("We are 95% confident that the population mean lies within this interval.")
```

```python
# Q29: Create a Python function that calculates and visualizes Type 1 and Type 2 errors during hypothesis testing
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def plot_type_errors(mu0=0, mu1=1, sigma=1, alpha=0.05, n=30):
    se = sigma / np.sqrt(n)
    critical_value = norm.ppf(1 - alpha)
    x = np.linspace(-4, 4, 1000)

    plt.plot(x, norm.pdf(x, mu0, se), label='H0 Distribution')
    plt.plot(x, norm.pdf(x, mu1, se), label='H1 Distribution')

    plt.fill_between(x, norm.pdf(x, mu0, se), where=x > critical_value, color='red', alpha=0.5, label='Type I Error')
    plt.fill_between(x, norm.pdf(x, mu1, se), where=x < critical_value, color='blue', alpha=0.3, label='Type II Error')

    plt.axvline(critical_value, color='black', linestyle='--', label='Critical Value')
    plt.title("Type I and Type II Errors")
    plt.xlabel("Test Statistic")
    plt.ylabel("Probability Density")
    plt.legend()
    plt.grid(True)
    plt.show()

plot_type_errors()
```

```python
# Q33: Write a Python function to calculate the confidence interval for a sample mean and explain its significance
import numpy as np
import scipy.stats as stats

def get_confidence_interval(data, confidence=0.95):
    mean = np.mean(data)
    sem = stats.sem(data)
    margin = sem * stats.t.ppf((1 + confidence) / 2, len(data)-1)
    return mean - margin, mean + margin

data = np.random.normal(50, 10, 40)
ci = get_confidence_interval(data)
print(f"Confidence interval: {ci}")
print("This interval represents the range in which the true mean likely lies with 95% confidence.")
```

```python
# Q35: Implement a Bayesian inference method using Bayes' Theorem in Python and explain the process
def bayes_theorem(prior_A, likelihood_B_given_A, prob_B):
    return (likelihood_B_given_A * prior_A) / prob_B

prior_A = 0.01           # Prior probability of disease
likelihood_B_given_A = 0.95  # Sensitivity
false_positive = 0.05
prob_B = (prior_A * likelihood_B_given_A) + ((1 - prior_A) * false_positive)

posterior = bayes_theorem(prior_A, likelihood_B_given_A, prob_B)
print(f"Posterior probability (P(A|B)): {posterior:.4f}")
print("Interpretation: Probability that a person has the disease given they tested positive.")
```
